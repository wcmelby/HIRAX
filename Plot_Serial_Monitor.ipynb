```python
import pandas as pd
import matplotlib.pyplot as plt
```

Plot the csv file data from reading the Arduino serial monitor for HIRAX PID temperature control. 

```python
# Read the CSV file
data = pd.read_csv('./Serial_Monitor_Files/Test1.csv')

# Plot the data
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(data['Time (s)'], data['% Power Output'], label='% Power Output', color='red')
plt.xlabel('Time (s)', fontweight='bold', fontsize='12')
plt.ylabel('% Power Output', fontweight='bold', fontsize='12')
plt.legend(fontsize='11')

plt.subplot(2, 1, 2)
plt.plot(data['Time (s)'], data['RTD Temp1 (C)'], label='RTD 1 (center)', color='forestgreen')
plt.plot(data['Time (s)'], data['RTD Temp2 (C)'], label='RTD 2 (rim)', color='orange')
plt.plot(data['Time (s)'], data['Time (s)']/data['Time (s)']*30, label='Setpoint', linestyle='--', color='gray')
plt.xlabel('Time (s)', fontweight='bold', fontsize='12')
plt.ylabel('Temperature (C)', fontweight='bold', fontsize='12')
plt.legend(fontsize='11')

plt.tight_layout()
plt.show()
```
