```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
import numpy as np
```

Plot the csv file data from reading the Arduino serial monitor for HIRAX PID temperature control. 

```python
# Read the CSV file
data = pd.read_csv('./Serial_Monitor_Files/Test1.csv')

# Plot the data
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(data['Time (s)'], data['% Power Output'], label='% Power Output', color='red')
plt.xlabel('Time (s)', fontweight='bold', fontsize='12')
plt.ylabel('% Power Output', fontweight='bold', fontsize='12')
plt.legend(fontsize='11')

plt.subplot(2, 1, 2)
plt.plot(data['Time (s)'], data['RTD Temp1 (C)'], label='RTD 1 (center)', color='forestgreen')
plt.plot(data['Time (s)'], data['RTD Temp2 (C)'], label='RTD 2 (rim)', color='orange')
plt.plot(data['Time (s)'], data['Time (s)']/data['Time (s)']*30, label='Setpoint', linestyle='--', color='gray')
plt.xlabel('Time (s)', fontweight='bold', fontsize='12')
plt.ylabel('Temperature (C)', fontweight='bold', fontsize='12')
plt.legend(fontsize='11')

plt.tight_layout()
plt.show()
```

```python
def analyze_temperature(csv_file):
    # Step 1: Read the CSV file and load the data into a list
    data = pd.read_csv(csv_file)
    time = data['Time (s)']
    temp1 = data['RTD Temp1 (C)']

    # Step 2: Chunk the data into sections of 30 data points each after temperature stabilizes
    chunk_size = 30
    # temp1_chunks = [temp1[i:i + chunk_size] for i in range(145, len(data), chunk_size)] # start after 75 seconds
    temp1_chunks = [temp1[i:i + chunk_size] for i in range(411, len(data), chunk_size)] # start after 200 seconds


    average_temperatures = []
    std_deviations = []

    for chunk in temp1_chunks:
        average_temperatures.append(np.mean(chunk))
        std_deviations.append(np.std(chunk))

    # for i, avg_temp in enumerate(average_temperatures):
    #     print(f"Average temperature for chunk {i + 1}: {avg_temp}, Standard deviation: {std_deviations[i]}")

    overall_avg_temp = np.mean(average_temperatures)
    overall_avg_std_deviation = np.mean(std_deviations) # Average standard deviation in a given 15 sec interval
 
    return average_temperatures, std_deviations, overall_avg_temp, overall_avg_std_deviation
```

```python
# 90, 3.3, 0.01
average_temperatures, std_deviations, overall_avg_temp, overall_std_deviation = analyze_temperature('./Serial_Monitor_Files/Test13.csv')
print(f"Overall average temperature: {overall_avg_temp}, Overall average standard deviation: {overall_std_deviation}")
```
